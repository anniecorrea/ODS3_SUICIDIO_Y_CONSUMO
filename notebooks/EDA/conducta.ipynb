```python
# ========= 0) Imports & Paths =========
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------ RUTA ------------
SUI_CSV = Path("../data/conductasuicida.csv")  # <-- AJUSTA
print("Existe el archivo?", SUI_CSV.exists())

# ========= 1) Carga =========
sui = pd.read_csv(SUI_CSV)
print("SHAPE:", sui.shape)
display(sui.head(5))
sui.info()
display(sui.describe(include="all").T.head(30))

# ========= 2) Nulos por columna (suma) & únicos =========
print("\nNULOS POR COLUMNA (suma)")
display(sui.isna().sum().sort_values(ascending=False))

print("\nNÚMERO DE ÚNICOS POR COLUMNA")
display(sui.nunique())

# ========= 3) Normalización ligera & set 'sin dato' =========
bad_set = {"", "nan", "sin dato", "upz sin asignar", "s/d", "n.a.", "n.a", "na"}
sui["nombre_upz"] = sui["nombre_upz"].astype("string").str.strip()
sui["localidad_residencia"] = sui["localidad_residencia"].astype("string").str.strip()

# ========= 3.1) Unificar variantes de “sin dato” =========
su = sui["nombre_upz"].str.lower()
mask_bad_upz = su.str.match(r'^(sin dato|upz sin asignar|n\.?a\.?|n/?a|s/?d|na)$', na=True)
sui["UPZ_norm"] = su.where(~mask_bad_upz, "sin dato").fillna("sin dato")

sl = sui["localidad_residencia"].str.lower()
mask_bad_loc = sl.str.match(r'^(sin dato|n\.?a\.?|n/?a|s/?d|na)$', na=True)
sui["LOC_norm"] = sl.where(~mask_bad_loc, "sin dato").fillna("sin dato")

# ========= 4) Copia para comparación (ANTES) =========
upz_before_sui = sui["nombre_upz"].copy()
upz_before_norm_sui = sui["UPZ_norm"].copy()

# ========= 5–6) Imputación UPZ por moda (con métricas) =========
bad_text_mask_pre = sui["nombre_upz"].astype("string").str.strip().str.lower().isin(bad_set)
print("Antes - filas con UPZ sin dato:", int(bad_text_mask_pre.sum()))

m_bad_upz = sui["nombre_upz"].astype("string").str.strip().str.lower().isin(bad_set) | sui["nombre_upz"].isna()
m_bad_loc = sui["localidad_residencia"].astype("string").str.strip().str.lower().isin(bad_set) | sui["localidad_residencia"].isna()

tmp = sui.loc[~m_bad_upz & ~m_bad_loc, ["localidad_residencia","nombre_upz"]].copy()
modo_por_loc = tmp.groupby("localidad_residencia")["nombre_upz"] \
                  .agg(lambda s: s.mode().iloc[0] if not s.mode().empty else np.nan)

sui.loc[m_bad_upz, "nombre_upz"] = sui.loc[m_bad_upz, "localidad_residencia"].map(modo_por_loc)

bad_text_mask_post = sui["nombre_upz"].astype("string").str.strip().str.lower().isin(bad_set)
print("Imputadas:", int(bad_text_mask_pre.sum() - bad_text_mask_post.sum()))
print("Aún sin UPZ (p.ej. localidad sin dato o sin moda):", int(bad_text_mask_post.sum()))

bad_with_nan_post = bad_text_mask_post | sui["nombre_upz"].isna()
print("Aún sin UPZ (incluye NaN):", int(bad_with_nan_post.sum()))

# ========= 6bis) Series normalizadas DESPUÉS =========
su_after = sui["nombre_upz"].astype("string").str.strip().str.lower()
mask_bad_after = su_after.str.match(r'^(sin dato|upz sin asignar|n\.?a\.?|n/?a|s/?d|na)$', na=True)
upz_after_norm_sui = su_after.where(~mask_bad_after, "sin dato").fillna("sin dato")

# ========= 7) Gráficas ANTES vs DESPUÉS =========
plt.figure(figsize=(9,4))
upz_before_norm_sui.value_counts().head(10).plot(kind="bar")
plt.title("Suicidas · Top 10 UPZ ANTES (normalizado)"); plt.ylabel("Casos"); plt.xlabel("UPZ")
plt.tight_layout(); plt.show()

plt.figure(figsize=(9,4))
upz_after_norm_sui.value_counts().head(10).plot(kind="bar")
plt.title("Suicidas · Top 10 UPZ DESPUÉS (normalizado)"); plt.ylabel("Casos"); plt.xlabel("UPZ")
plt.tight_layout(); plt.show()

# ========= 8) Duplicados por clave de merge =========
key_sui = [c for c in ["ano_notificacion","nombre_upz","sexo","ciclovital","niveleducativo"] if c in sui.columns]
print("\nCLAVE SUI:", key_sui)
if key_sui:
    dup_count = sui.duplicated(subset=key_sui, keep=False).sum()
    print("Duplicados por clave:", dup_count)
    if dup_count > 0:
        display(sui.loc[sui.duplicated(subset=key_sui, keep=False), key_sui].head(10))

# ========= 9) Frecuencias completas de UPZ (normalizado) =========
sui_upz_freq = upz_after_norm_sui.value_counts().rename_axis("UPZ").reset_index(name="frecuencia")
print("\nSuicidas – UPZ (todas) con frecuencias (normalizado):")
display(sui_upz_freq)
# sui_upz_freq.to_csv("suicidas_upz_frecuencias.csv", index=False)
```
